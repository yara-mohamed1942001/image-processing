```python
import numpy as np
import cv2

# Load the image
#image = cv2.imread('your_image.jpg')

# Define mean and standard deviation for Gaussian noise
mu = 0
sigma = 25

# Generate Gaussian noise
noise = np.random.normal(mu, sigma, imggray.shape).astype(np.uint8)

# Add noise to the image
noisy_image = imggray+ noise


plt.imshow(noisy_image,cmap=plt.get_cmap('gray'))
plt.show()
```

```python
blurred_image = cv2.blur(noisy_image, (5,5))
plt.imshow(blurred_image,cmap=plt.get_cmap('gray'))
plt.show()
```

```python
blurred_image = cv2.blur(noisy_image, (8,8))
plt.imshow(blurred_image,cmap=plt.get_cmap('gray'))
plt.show()
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('C://Users//owner//Pictures//Saved Pictures//pexels-pixabay-60597.jpg')

# Add Gaussian noise to the image
mean = 0
sigma = 25
gaussian_noise = np.random.normal(mean, sigma, image.shape).astype(np.uint8)
noisy_image = cv2.add(image, gaussian_noise)
noisy_image = np.clip(noisy_image, 0, 255)

# Convert the noisy image to grayscale
gray_image = cv2.cvtColor(noisy_image, cv2.COLOR_BGR2GRAY)

# Define the kernel size for the median filter
kernel_size = 5  # Adjust the size as needed (should be odd)

# Apply the median filter to the grayscale image
filtered_image = cv2.medianBlur(gray_image, kernel_size)

plt.imshow(filtered_image,cmap=plt.get_cmap('gray'))
plt.show()
```
